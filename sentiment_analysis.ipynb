```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "task2.csv"   # make sure the file is in the same folder
df = pd.read_csv(file_path)

# Display basic info
print("ðŸ”¹ First 5 Rows:")
print(df.head())
print("\nðŸ”¹ Dataset Info:")
print(df.info())
print("\nðŸ”¹ Summary Statistics:")
print(df.describe())

# Check for missing values
print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())

# Handle missing values (optional)
df = df.dropna()  # or df.fillna(method='ffill') to forward fill

# Check for duplicates
print("\nðŸ”¹ Duplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()

# Plot distributions of numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols].hist(figsize=(10, 8), bins=15)
plt.suptitle("Distribution of Numerical Columns")
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Countplot for categorical columns
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col)
    plt.title(f"Countplot of {col}")
    plt.xticks(rotation=45)
    plt.show()

print("\nâœ… Data Cleaning and EDA Completed Successfully!")
```
